```python
import json

mock_json_response = """
{
    "bpi": {
        "USD": {
            "code": "USD",
            "rate": "64,250.50",
            "description": "United States Dollar"
        }
    }
}
"""

data = json.loads(mock_json_response)

bitcoin_price = data["bpi"]["USD"]["rate"]
print(f"current price : {bitcoin_price}")
```
